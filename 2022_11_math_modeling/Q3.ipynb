```python
import pandas as pd
import numpy as np
import math
from numpy import array


# 定义熵值法函数
def cal_weight(x):
    x = x.apply(lambda x: ((x - np.min(x)) / (np.max(x) - np.min(x))))
    rows = x.index.size 
    cols = x.columns.size 
    k = 1.0 / math.log(rows)
    lnf = [[None] * cols for i in range(rows)]
    x = array(x)
    lnf = [[None] * cols for i in range(rows)]
    lnf = array(lnf)
    for i in range(0, rows):
        for j in range(0, cols):
            if x[i][j] == 0:
                lnfij = 0.0
            else:
                p = x[i][j] / x.sum(axis=0)[j]
                lnfij = math.log(p) * p * (-k)
            lnf[i][j] = lnfij
    lnf = pd.DataFrame(lnf)
    E = lnf

    d = 1 - E.sum(axis=0)
    w = [[None] * 1 for i in range(cols)]
    for j in range(0, cols):
        wj = d[j] / sum(d)
        w[j] = wj
    w = pd.DataFrame(w)
    return w
```

要计算夏普比率，首先要知道指数的回报率，即当天指数减去前一天指数，得到的差值与前一天指数的比值。

```python
# 计算每nian指数回报率
index_returns = temp[temp.columns[1:]].pct_change()

# 回报率绘图
index_returns.plot()

```

```python
# 回报率统计量
index_returns.describe()
```

无风险收益率的年数据  3%

sharp-ratio里一定要注意分子分母周期的一致性。 一般国内无风年利率普遍习采用一年期存款利率，约为3%，




```python
#计算收益率的标准差
nn=index_returns.std()
nn
```

（4）计算夏普率结果

```python
index_returns['中证1000_夏普比率']=index_returns['中证1000'].apply(lambda x:(x-0.03)/nn['中证1000'])
index_returns['南华商品指数_夏普比率']=index_returns['南华商品指数'].apply(lambda x:(x-0.03)/nn['南华商品指数'])
index_returns['中债-综合财富(7-10年)指数_夏普比率']=index_returns['中债-综合财富(7-10年)指数'].apply(lambda x:(x-0.03)/nn['中债-综合财富(7-10年)指数'])
index_returns['货币基金_夏普比率']=index_returns['货币基金'].apply(lambda x:(x-0.03)/nn['货币基金'])
index_returns
```
